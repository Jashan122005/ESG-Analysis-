# Cleaning the data

```python
# Step 1: Import required libraries
import pandas as pd      
import numpy as np       
import matplotlib.pyplot as plt  
import seaborn as sns    

df = pd.read_csv("ESG Data.csv")

print(df.head())

print(df.info())
```

```python
# Convert all industry values to uppercase and strip spaces
df['industry'] = df['industry'].astype(str).str.strip().str.upper()

# Remove rows where industry is "N/A" or "NA"
df_cleaned = df[df['industry'] != "NaN"]
df_cleaned = df_cleaned[df_cleaned['industry'] != "NAN"]

print("Before:", df.shape)
print("After :", df_cleaned.shape)
```

```python
df_cleaned['logo'] = df_cleaned['logo'].astype(str).str.strip().str.upper()

# Remove rows where industry is "N/A" or "NA"
df_clean = df_cleaned[df_cleaned['logo'] != "NaN"]
df_clean = df_cleaned[df_cleaned['logo'] != "NAN"]

print("Before:", df.shape)
df_clean.shape
print("After :", df_clean.shape)
```

# Exploration Begins

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_esg_radar(row):
    labels = ['Environment', 'Social', 'Governance']
    scores = [row['environment_score'], row['social_score'], row['governance_score']]

    angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist()
    scores += scores[:1]
    angles += angles[:1]

    fig, ax = plt.subplots(figsize=(5,5), subplot_kw=dict(polar=True))
    ax.plot(angles, scores, 'o-', linewidth=2)
    ax.fill(angles, scores, alpha=0.25)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels)
    ax.set_title(row['name'])
    plt.show()

# Example: radar for first company
plot_esg_radar(df_clean.iloc[2])
```

```python
plt.figure(figsize=(12,6))
sns.stripplot(x="industry", y="total_score", data=df, jitter=True, palette="viridis", alpha=0.7)
plt.xticks(rotation=90)
plt.title("Distribution of ESG Scores by Industry", fontsize=14, fontweight='bold')
plt.show()
```

```python
corr = df[['environment_score','social_score','governance_score','total_score']].corr()

plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap="crest", vmin=-1, vmax=1, linewidths=0.5)
plt.title("Correlation Between ESG Dimensions", fontsize=14, fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.scatterplot(x="environment_score", y="social_score", hue="industry", 
                size="governance_score", sizes=(20,200), alpha=0.7, data=df, palette="tab20")
plt.title("ESG Score Relationships Across Industries", fontsize=14, fontweight='bold')
plt.show()
```

```python
top10 = df_clean.nlargest(10, 'total_score')[['name','total_score']]

plt.figure(figsize=(10,6))
sns.barplot(x="total_score", y="name", data=top10, palette="mako")
plt.title("Top 10 Companies by ESG Score", fontsize=14, fontweight='bold')
plt.xlabel("Total ESG Score")
plt.ylabel("Company")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df_clean, x='total_grade', order=df['total_grade'].value_counts().index,
              palette="coolwarm")
plt.title("Distribution of ESG Grades", fontsize=14, fontweight='bold')
plt.xlabel("ESG Grade")
plt.ylabel("Number of Companies")
plt.show()
```

```python
pivot_table = df.pivot_table(index='industry', columns='total_grade', values='ticker', aggfunc='count')

plt.figure(figsize=(12,6))
sns.heatmap(pivot_table, cmap="YlGnBu", annot=True, fmt="g")
plt.title("Industry vs ESG Grade Distribution", fontsize=14, fontweight='bold')
plt.xlabel("ESG Grade")
plt.ylabel("Industry")
plt.show()
```

```python
df_clean['last_processing_date'] = pd.to_datetime(df_clean['last_processing_date'], errors='coerce')

plt.figure(figsize=(10,5))
df_clean['last_processing_date'].dt.to_period("M").value_counts().sort_index().plot(kind='bar', color='teal')
plt.title("Number of ESG Assessments Over Time", fontsize=14, fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Assessments")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='industry', hue='total_level', palette="Set2")
plt.xticks(rotation=90)
plt.title("ESG Level Distribution by Industry", fontsize=14, fontweight='bold')
plt.xlabel("Industry")
plt.ylabel("Count")
plt.legend(title="ESG Level")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='environment_score', y='social_score', size='governance_score',
                hue='total_grade', alpha=0.7, data=df, palette="Spectral", sizes=(50,300))
plt.title("ESG Dimensions with Grade Coloring", fontsize=14, fontweight='bold')
plt.xlabel("Environment Score")
plt.ylabel("Social Score")
plt.show()
```

```python
currency_counts = df_clean['currency'].value_counts(normalize=True).mul(100).sort_values()

plt.figure(figsize=(8,5))
sns.barplot(x=currency_counts, y=currency_counts.index, palette="viridis")
plt.title("Percentage of Companies by Currency", fontsize=14, fontweight="bold")
plt.xlabel("Percentage (%)")
plt.ylabel("Currency")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count of companies grouped by Industry and Stock Exchange
industry_exchange = df_clean.groupby(["industry", "exchange"]).size().reset_index(name="count")

plt.figure(figsize=(18,8))
sns.barplot(
    data=industry_exchange,
    x="industry",
    y="count",
    hue="exchange",
    palette="tab10"
)
plt.title("Distribution of Industries across Stock Exchanges", fontsize=14, fontweight="bold")
plt.xlabel("Industry")
plt.ylabel("Number of Companies")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Stock Exchange", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
